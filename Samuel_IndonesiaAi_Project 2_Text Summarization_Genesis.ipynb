```python
import os
import json
import pandas as pd

# Path folder
train_dir = "/content/liputan6_data/liputan6_data/canonical/train"
dev_dir = "/content/liputan6_data/liputan6_data/canonical/dev"
test_dir = "/content/liputan6_data/liputan6_data/canonical/test"

def load_selected_fields(directory, split_name):
    data = []
    for filename in os.listdir(directory):
        if filename.endswith(".json"):
            file_path = os.path.join(directory, filename)
            try:
                with open(file_path, 'r') as file:
                    json_data = json.load(file)
                    data.append({
                        "id": json_data.get("id", ""),
                        "url": json_data.get("url", ""),
                        "clean_article": json.dumps(json_data.get("clean_article", []), ensure_ascii=False),
                        "clean_summary": json.dumps(json_data.get("clean_summary", []), ensure_ascii=False),
                        "extractive_summary": json.dumps(json_data.get("extractive_summary", []), ensure_ascii=False),
                        "split": split_name
                    })
            except Exception as e:
                print(f"Gagal membaca file: {file_path} - {e}")
    return data

# Load masing-masing data
train_data = load_selected_fields(train_dir, "train")
dev_data = load_selected_fields(dev_dir, "dev")
test_data = load_selected_fields(test_dir, "test")

# Simpan ke tiga file CSV terpisah
pd.DataFrame(train_data).to_csv("liputan6_train.csv", index=False)
pd.DataFrame(dev_data).to_csv("liputan6_dev.csv", index=False)
pd.DataFrame(test_data).to_csv("liputan6_test.csv", index=False)

print(" CSV 'liputan6_train.csv', 'liputan6_dev.csv', dan 'liputan6_test.csv' berhasil dibuat.")
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

model_name = "cahya/t5-base-indonesian-summarization-cased"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)
```

```python
import json

def flatten_article(raw_json):
    parsed = json.loads(raw_json)
    tokens = [t for sent in parsed for t in sent]
    return " ".join(tokens[:512])

# Ambil contoh
sample_input = flatten_article(data.loc[0, "clean_article"])
sample_ref = flatten_article(data.loc[0, "clean_summary"])

input_ids = tokenizer.encode(sample_input, return_tensors="pt", truncation=True, max_length=512)
summary_ids = model.generate(
    input_ids,
    min_length=20,
    max_length=100,
    num_beams=5,
    no_repeat_ngram_size=2,
    early_stopping=True
)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
```

```python
import json
from tqdm import tqdm
import evaluate

rouge = evaluate.load("rouge")

def flatten_article(raw_json):
    try:
        parsed = json.loads(raw_json)
        tokens = [t for sent in parsed for t in sent]
        return " ".join(tokens[:512])
    except:
        return ""
```

```python
predictions = []
references = []

for i in tqdm(range(100)):
    article = flatten_article(data.loc[i, "clean_article"])
    summary_ref = flatten_article(data.loc[i, "clean_summary"])

    if not article or not summary_ref:
        continue

    input_ids = tokenizer.encode(article, return_tensors="pt", truncation=True, max_length=512)
    output_ids = model.generate(
        input_ids,
        min_length=20,
        max_length=100,
        num_beams=4,
        no_repeat_ngram_size=2,
        early_stopping=True
    )

    summary_pred = tokenizer.decode(output_ids[0], skip_special_tokens=True)

    predictions.append(summary_pred)
    references.append(summary_ref)
```

```python
results = rouge.compute(predictions=predictions, references=references)
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```

```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import json

# Load model dan tokenizer
model_name = "google/mt5-small"  # Bisa diganti nanti ke model yang sudah fine-tuned
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)

def flatten_article(article_json):
    try:
        parsed = json.loads(article_json)
        flattened = [token for sentence in parsed for token in sentence]
        return " ".join(flattened)
    except Exception as e:
        return f"[Error parsing input JSON]: {e}"

def summarize(article_json):
    article_text = flatten_article(article_json)
    if article_text.startswith("[Error"):
        return article_text

    inputs = tokenizer(
        article_text,
        return_tensors="pt",
        max_length=512,
        truncation=True
    )

    output_ids = model.generate(
        inputs.input_ids,
        max_new_tokens=100,
        num_beams=4,
        early_stopping=True
    )

    summary = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    return summary

# Gradio Interface
interface = gr.Interface(
    fn=summarize,
    inputs=gr.Textbox(label="Masukkan JSON 'clean_article' dari Liputan6", lines=15, placeholder="Masukkan JSON array kalimat di sini"),
    outputs=gr.Textbox(label="Ringkasan oleh Model"),
    title="Text Summarization Liputan6",
    description="Masukkan satu artikel dalam bentuk JSON dari dataset Liputan6, dan model akan mencoba merangkum."
)

interface.launch(debug=True)
```
